```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def get_line_function(m = 0, b = 0):
    return lambda x: m*x + b
```

```python
f = get_line_function(m = 1, b = 0)
```

```python
def heaviside(x):
    if x<0:
        return -1
    elif x>0:
        return 1
    else:
        return 0
```

```python
def signum(x):
    if x<0:
        return 0
    elif x>0:
        return 1
    else:
        return 0.5
```

```python
def piece_wise(x):
    if x<=-0.5:
        return 0
    elif -0.5<x<0.5:
        return x+0.5
    else:
        return 1
```

```python
def sigmoid(x):
    return 1/(1+np.e**-x)
```

```python
def htan(x):
    return (np.e**x - np.e**-x)/(np.e**x + np.e**-x)
```

```python
def relu(x):
    return 0 if x<=0 else x
```

```python
fig, ax = plt.subplots(figsize = [8,8])

fig.suptitle('f(x) = x²', fontsize = 26)

ax.set_aspect(1)

ax.set_xmargin(0)
ax.set_ymargin(0)
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

ax.grid(zorder = 1)
ax.axhline(0, color = 'black', zorder = 2)
ax.axvline(0, color = 'black', zorder = 2)

X = np.linspace(-5, 5, 100)

# f(x) = x²
f = lambda x: x**2

ax.plot(X, f(X), zorder = 3, label = 'f(x)')
ax.plot(X, f(X)+1, zorder = 3, label = 'f(x)+1')
ax.plot(X, f(X)-1, zorder = 3, label = 'f(x)-1')
ax.plot(X, f(X+1), zorder = 3, label = 'f(x+1)')
ax.plot(X, f(X-1), zorder = 3, label = 'f(x-1)')
ax.plot(X, f(X)*3, zorder = 3, label = 'f(x)*3')
ax.plot(X, f(X*3), zorder = 3, label = 'f(x*3)')
ax.plot(X, f(X)/3, zorder = 3, label = 'f(x)/3')
ax.plot(X, f(X/3), zorder = 3, label = 'f(x/3)')
ax.plot(X, -f(X), zorder = 3, label = '-f(x)')
ax.plot(X, f(-X), zorder = 4, label = 'f(-x)', linestyle = ':', linewidth = '4', color = 'black')

ax.legend(fontsize = 'large')

fig.savefig('./fig1.png')
```

```python
fig, ax = plt.subplots(figsize = [15,4])

fig.suptitle('f(x) = x²', fontsize = 26)

ax.set_aspect(1)

ax.set_xmargin(0)
#ax.set_ymargin(0)
#ax.set_xlim(-2, 2)
#ax.set_ylim(-2, 2)

ax.grid(zorder = 1)
ax.axhline(0, color = 'black', zorder = 2)
ax.axvline(0, color = 'black', zorder = 2)

X = np.linspace(-5, 5, 1000)

f = lambda x: x**2
g = lambda x: np.sin(x)

ax.plot(X, g(f(X)), zorder = 3, label = 'g(f(x))')
ax.plot(X, f(g(X)), zorder = 3, label = 'f(g(x))')

ax.legend(fontsize = 'xx-large')
```

```python
fig, ax = plt.subplots(figsize = [16, 8])

fig.suptitle('Funciones de activación', fontsize = 26)

ax.set_aspect(1, )

ax.set_xmargin(0)
#ax.set_ymargin(0)
#ax.set_xlim(-2, 2)
ax.set_ylim(-1.5, 1.5)

ax.grid(zorder = 1)
ax.axhline(0, color = 'grey', zorder = 2)
ax.axvline(0, color = 'grey', zorder = 2)

X = np.linspace(-3, 3, 1000)

ax.plot(X, X, zorder = 3, label = 'linear')
ax.plot(X, list(map(piece_wise, X)), zorder = 3, label = 'piece wise')
ax.plot(X, list(map(unit_step, X)), zorder = 3, label = 'unit step')
ax.plot(X, list(map(signum, X)), zorder = 3, label = 'signum')
ax.plot(X, list(map(sigmoid, X)), zorder = 3, label = 'sigmoid')
ax.plot(X, list(map(htan, X)), zorder = 3, label = 'hyperbolic tangent')
ax.plot(X, list(map(relu, X)), zorder = 4, label = 'relu')


ax.legend(fontsize = 'xx-large')

fig.savefig('./fig2.png')
```

# Limites

## Límite lateral izquierdo

$\lim_{x \to a^-} f(x) = L$

## Límite lateral derecho

$\lim_{x \to a^+} f(x) = L$


## Limites de f(x) = 1/x cuando x-> 0

```python
from sympy import limit, symbols
x = symbols('x')
```

```python
l = limit(1/x, x, 0, '+')
print('Limite derecho')
l
```

```python
l = limit(1/x, x, 0, '-')
print('limite izquierdo')
l
```
