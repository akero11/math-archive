```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def get_line_function(m = 0, b = 0):
    return lambda x: m*x + b
```

```python
f = get_line_function(m = 1, b = 0)
```

```python
def heaviside(x):
    if x<0:
        return -1
    elif x>0:
        return 1
    else:
        return 0
```

```python
def signum(x):
    if x<0:
        return 0
    elif x>0:
        return 1
    else:
        return 0.5
```

```python
def piece_wise(x):
    if x<=-0.5:
        return 0
    elif -0.5<x<0.5:
        return x+0.5
    else:
        return 1
```

```python
def sigmoid(x):
    return 1/(1+np.e**-x)
```

```python
def htan(x):
    return (np.e**x - np.e**-x)/(np.e**x + np.e**-x)
```

```python
def relu(x):
    return 0 if x<=0 else x
```

```python
fig, ax = plt.subplots(figsize = [8,8])

ax.set_aspect(1)

ax.set_xmargin(0)
ax.set_ymargin(0)
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

ax.grid(zorder = 1)
ax.axhline(0, color = 'black', zorder = 2)
ax.axvline(0, color = 'black', zorder = 2)

# Infinite line
X = (-5, 5)
P = ((x, f(x)) for x in X)
ax.axline(*P, color = 'orange', zorder = 3, label = 'line')

X = np.linspace(-5, 5, 100)

#x²
Y = X**2
ax.plot(X, Y, zorder = 4, label = 'x²')

# sin(x)
Y = np.sin(X)
ax.plot(X, Y, zorder = 5, label = 'sin')

# cos(x)
Y = np.cos(X)
ax.plot(X, Y, zorder = 6, label = 'cos')

# heaviside
Y = [heaviside(x) for x in X]
ax.plot(X, Y, zorder = 7, label = 'Heaviside', linewidth = '3')

Y = [signum(x) for x in X]
ax.plot(X, Y, zorder = 8, label = 'Signum', linestyle = '--', linewidth = '3', color = 'purple')

ax.legend(fontsize = 'x-large')

```

```python
from matplotlib import cm
```

```python
def sincos(x, y):
    return np.sin(x) + np.cos(y)

res = 100

x = np.linspace(-4, 4, num = res)
y = np.linspace(-4, 4, num = res)

x, y = np.meshgrid(x, y)

z = sincos(x, y)
```

```python
fig , ax = plt.subplots(subplot_kw = {'projection': '3d'}, figsize = [20, 10])

fig.suptitle('z = sin(x) + cos(y)', fontsize = 26)

surf = ax.plot_surface(x, y, z, cmap = cm.hot)

fig.colorbar(surf)
```

```python
fig2, ax2 = plt.subplots()
ax2.set_aspect(1)

level_map = np.linspace(np.min(z), np.max(z), num = res)

cp = ax2.contourf(x, y, z, levels = level_map, cmap = cm.hot)

fig2.colorbar(cp)
```

```python
def f(x, y):
    return x**2 + y**2

res = 100

x = np.linspace(-4, 4, num = res)
y = np.linspace(-4, 4, num = res)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

```python
fig, ax = plt.subplots(subplot_kw = {'projection': '3d'}, figsize = [20, 10])

fig.suptitle('z = x² + y²', fontsize = 26)

surf = ax.plot_surface(X, Y, Z, cmap = cm.hot)

fig.colorbar(surf)
```

```python
h = 0.01
lr = 0.01

def derivate(cp, p):
    return (f(cp[0], cp[1]) - f(p[0], p[1]))/h

def gradient(p):
    grad = np.zeros(2)
    for idx, val in enumerate(p):
        cp = np.copy(p)
        cp[idx] = cp[idx] + h
        
        dp = derivate(cp, p)
        
        grad[idx] = dp 
        
    return grad
```

```python
fig2, ax2 = plt.subplots()
fig2.suptitle('Descenso del gradiente', fontsize = 26)
ax2.set_aspect(1)
level_map = np.linspace(np.min(Z), np.max(Z), num = res)
cp = ax2.contourf(X, Y, Z, levels = level_map, cmap = cm.hot)
fig2.colorbar(cp)

p = np.random.rand(2) * 8 - 4
ax2.plot(p[0], p[1], 'o', c = 'blue')

for i in range(100):
    p = p - lr*gradient(p)
    if (i%10==0):
        ax2.plot(p[0], p[1], 'o', c = 'green')

ax2.plot(p[0], p[1], 'o', c = 'yellow')
```
