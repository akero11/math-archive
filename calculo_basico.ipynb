```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def get_line_function(m = 0, b = 0):
    return lambda x: m*x + b
```

```python
f = get_line_function(m = 1, b = 0)
```

```python
fig, ax = plt.subplots(figsize = [6,6])

ax.set_aspect(1, adjustable = 'datalim')

ax.set_xmargin(0)
ax.set_ymargin(0)

ax.grid(zorder = 1)
ax.axhline(0, color = 'black', zorder = 2)
ax.axvline(0, color = 'black', zorder = 2)

# Infinite line
X = (-5, 5)
P = ((x, f(x)) for x in X)
ax.axline(*P, color = 'orange', zorder = 3)

# x²
X = np.linspace(-5, 5, 100)
Y = X**2
ax.plot(X, Y, zorder = 4)
```
